```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()


x = np.random.rand(50)
y = 2 * x + np.random.randn(50) * 0.1
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.show()

time = np.arange(0, 10, 0.1)
amplitude = np.sin(time)
plt.plot(time, amplitude)
plt.title("Line Plot")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()





```
